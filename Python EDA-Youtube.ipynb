```python
def fill_numeric(data):
    for i in data.columns:
        data[i]=data[i].fillna(data[i].median())
    print(data.isnull().sum())
    return data
```

```python
def fill_categoric(data):
    for i in data.columns:
        data[i]=data[i].fillna(method='ffill')
    print(data.isnull().sum())
    return data
```

```python
filled_numeric=fill_numeric(new_numeric_data)
filled_numeric
```

```python
filled_category=fill_categoric(new_category_data)
```

```python
#4. problem outliers
from scipy.stats  import zscore
```

```python
def zscore_calci(data):
    for i in data.columns:
        data[i]=np.abs(zscore(data[i]))
    x=np.where(data>3)
    data.drop(data.index[x[:1]],inplace=True) # dropping it through rows
    return data
```

```python
zscore_data=zscore_calci(filled_numeric)
zscore_data
```

```python
new_reset_data = zscore_data.reset_index(drop=True)
new_reset_data
```
