```python
twitter = pd.read_csv(r'F:/Excel files/twitter acc/Top-1000-Celebrity-Twitter-Accounts.csv')
```

```python
twitter.drop(['twitter_userid','twitter_username','domain'],axis=1,inplace=True)
```

```python
twitter.dropna(subset=['followers_count','tweet_count'], how='all',inplace=True)
```

```python
df1=twitter.sort_values(by='followers_count',ascending=False)[:10]
df1
```

```python
plt.subplots(figsize=(18,9))
sns.barplot(y='name',x='followers_count',data=df1
            ,palette='husl',edgecolor=sns.color_palette('dark',10))
plt.xticks(rotation=90,weight='bold',fontsize=12)
plt.xlabel('Number of Followers',fontsize=15,weight='bold')
plt.ylabel('Person Name',fontsize=15,weight='bold')
plt.yticks(weight='bold',fontsize=12)
plt.title('Top 10 People with most followers',weight='bold',fontsize=20)
plt.show()
```

```python
plt.subplots(figsize=(18,9))
sns.barplot(y='name',x='tweet_count',data=df1
            ,palette='husl',edgecolor=sns.color_palette('dark',10))
plt.xticks(rotation=90,weight='bold',fontsize=12)
plt.xlabel('Number of tweet',fontsize=15,weight='bold')
plt.ylabel('Person Name',fontsize=15,weight='bold')
plt.yticks(weight='bold',fontsize=12)
plt.title('Top 10 People with most followers with Number of Tweets',weight='bold',fontsize=20)
plt.show()
```
